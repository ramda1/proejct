```python
import numpy as np
from sklearn.preprocessing import StandardScaler, RobustScaler
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#iqr 지정 함수

def outliers_iqr(data):

  q1, q3 = np.percentile(data, [25, 75])
  iqr = q3 - q1

  lower_bound = q1 - (iqr * 1.5)
  upper_bound = q3 + (iqr * 1.5)

  return np.where((data > upper_bound) | (data < lower_bound))
```

```python
outliers_na = outliers_iqr(df['NA_Sales'])[0]
outliers_jp = outliers_iqr(df['JP_Sales'])[0]
outliers_eu = outliers_iqr(df['EU_Sales'])[0]
outliers_ot = outliers_iqr(df['Other_Sales'])[0]
```

```python
df.loc[outliers_na, 'NA_Sales']
df.loc[outliers_jp, 'JP_Sales']
df.loc[outliers_eu, 'EU_Sales']
df.loc[outliers_ot, 'Other_Sales']
```

```python
outlier_all = np.concatenate((outliers_na, outliers_jp, outliers_eu, outliers_ot), axis = None)
print(len(outlier_all))
outlier_all
```

```python
outliers_n = []

for i in df.index:

  if i not in outlier_all:
    outliers_n.append(i)
```

```python
#북미지역의 장르별 출고량/ 이하 출고량 모두 million 기준

a = {}
uniq_genre = df_clean["Genre"].unique()

for Genre in uniq_genre:
  NA_Sales_sum = df_clean.loc[df_clean["Genre"] == Genre, "NA_Sales"].sum()
  a[Genre] = NA_Sales_sum

dataFrame = pd.DataFrame.from_dict([a])
dataFrame = dataFrame.T

dataFrame.columns = ['NA_Sales']
dataFrame

dataFrame.sort_values('NA_Sales', ascending = False)
```
