## column 용어 정리

```python
# column 용어 정리

# id : 중고차 회원(구매자)의 개별 id
# vin : 차대번호(사람으로 따지면 주민등록번호, 차량 절도 내역 조회 가능)
# price : 가격 // 이걸 target으로
# miles :  주행거리 / 배기량(차량이 낼 수 있는 동력 ) / rpm(바퀴가 회전하는 속도) / 연비 = 거리 / 연료량(1L당 갈 수 있는 거리) / 중고자체어슨
# stock_no : 웹사이트에 나와 있는? 차량의 재고 번호
# year : 연식
# make : 제조사
# model : 모델
# trim : 트림(모델 안에서의 등급) / http://wiki.hash.kr/index.php/%ED%8A%B8%EB%A6%BC
# body_type : 바디 스타일 / https://blog.gm-korea.co.kr/7
# vehicle_type : 차량 유형
# drivetrain : 구동 방식 / FWD : 전륜구동 자동차, RWD : 후륜구동 자동차, 4WD: 평소에는 두바퀴만 굴리는데(보통 후륜구동) 필요에 따라서 앞바퀴도 굴리게끔 전환 가능한 모드가 심어져 있는 자동차 / https://kids.hyundai.com/kidshyundai/AutomobileManagement/learnauto/automobileDet.kids?cotnSn=3550
# transmission : 변속기 종류
# fuel_type : 연료 타입
# engine_size : 엔진의 배기량(L 단위)
# engine_block : 실린더 블록(4개의 진원통형인 실린더보어는 내부에 투입되는 혼합기(공기 + 연료)의 폭발 에너지를 피스톤의 왕복 운동으로 변환하고
                              #다시 크랭크에서 왕복 운동을 회전 운동으로 변환시켜 자동차를 움직임) / http://www.maxrotec.com/content/product_view.php?uuid=5feddcec-9b67-493c-825b-ab637ee3079f&seq=30&pseq=30&no=13&PHPSESSID=7e7d995f4700c103bae8e0b906168f6a

                              # 직렬 엔진(I) : 준중형, 소형차에 많이 쓰임, V형 엔진 : 자동차에서 가장 일반적으로 쓰이는 엔진, H형 엔진 : 주로 항공기에 많이 쓰임

# 연도, 주행거리, 변속기 종류, 배기량, 연료타입, 구동방식 중요

# vin, id, seller_name, street, zip는 제외시킴
# city, state는 일단 보류
```

```python
import pandas as pd

ca = pd.read_csv('/content/drive/MyDrive/ca-dealers-used.csv')
dfc = ca.dropna()   # 결측치 제거


us = pd.read_csv('/content/drive/MyDrive/us-dealers-used.csv')
dfu = us.dropna()
```

## 타입 변경

```python
dfc[["year"]] = dfc[["year"]].astype("int")
dfu[["year"]] = dfu[["year"]].astype("int")
dfc.info(), dfu.info()
```

## 이상치 제거

```python
# Canada

q1 = dfc.quantile(0.25)
q3 = dfc.quantile(0.75)
IQR = q3 - q1
condition = (dfc < (q1 - 1.5 * IQR)) | (dfc > (q3 + 1.5 * IQR))
condition = condition.any(axis=1)
search_dfc = dfc[condition]

dfc = dfc.drop(search_dfc.index, axis=0)
dfc.info()
```

```python
# USA

q1 = dfu.quantile(0.25)
q3 = dfu.quantile(0.75)
IQR = q3 - q1
condition = (dfu < (q1 - 1.5 * IQR)) | (dfu > (q3 + 1.5 * IQR))
condition = condition.any(axis=1)
search_dfu = dfu[condition]

dfu = dfu.drop(search_dfu.index, axis=0)
dfu.info()
```
